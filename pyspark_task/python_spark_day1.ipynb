```
dfst = spark.read.csv("/FileStore/tables/example.csv", header = True, inferSchema=True)
```

```
data18 = dfst.filter(dfst['year'] == 2018)
data18 = data18.drop('year')
```

```
data19 = dfst.filter(dfst['year'] == 2019)
data19 = data19.drop('year')
data19 = data19.withColumnRenamed('adv','adv19')
data19 = data19.withColumnRenamed('channel','channel19')
```

```
df = data18.join(data19, (data18["adv"] == data19["adv19"]) & (data18['channel'] == data19['channel19']))
```

```
tempdf = data18.join(data19, (data18['channel']== data19['channel19'] ))
```

```
tempdf.join( df, 'adv','leftanti').join(df,'adv19' ,'leftanti').drop('channel19').select('channel','adv','adv19').distinct().show()
```
